```python
import pandas as pd
data_frame = pd.read_csv("data.csv",index_col=0)
df1 = pd.DataFrame(data_frame[["Name","Value","Wage"]])
df1
```

```python
def Money_Converter(money):
    if(type(money)==float or type(money)==int):
        return money
    value = 1
    if(money[0]=='€'):
        if(money[-1]=="K"):
            value = float(money[1:-1])*1000
            return value
        if(money[-1]=="M"):
            value = float(money[1:-1])*1000000
            return value
    return 0.0
Money_Converter("0")   
```

```python
value = df1["Value"].apply(Money_Converter)
wage = df1["Wage"].apply(Money_Converter)
df1["Value"] = value
df1["Wage"] = wage
df1["difference"] = df1["Value"] - df1["Wage"]
df1
```

```python
df1.sort_values("difference", ascending = False)
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(wage,Value)", "(@Wage, @Value)"),
    ("name","@Name")
])

p = figure(title = "Soccer 2020", x_axis_label = 'Wage', y_axis_label = 'Value', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source = df1)
show(p)
```
